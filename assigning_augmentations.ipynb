```python
import pandas as pd
import random
import numpy as np
```

```python
val=pd.read_csv('./val.csv')
train=pd.read_csv('./train.csv')
```

```python
val['Augmentation']=[0 for i in range(val.shape[0])]
val.head()
val.to_csv('./val_final.csv',index=False)
```

```python
p_val=val[val['Pneumothorax']==1]
np_val=val[val['Pneumothorax']==0]

p_train=train[train['Pneumothorax']==1]
np_train=train[train['Pneumothorax']==0]
```

```python
print("Train\nP : NP")
print(p_train.shape[0],":",np_train.shape[0],"=",p_train.shape[0]/np_train.shape[0])
print()
print("Validation\nP : NP")
print(p_val.shape[0],":",np_val.shape[0],"=",p_val.shape[0]/np_val.shape[0])
```

```python
shuffled=p_train.sample(frac=1)
```

```python
half=shuffled.shape[0]//2
thrice = shuffled.iloc[:half].loc[shuffled.index[:half].repeat(3)].reset_index(drop=True)
fours = shuffled.iloc[half:].loc[shuffled.index[half:].repeat(4)].reset_index(drop=True)
newdf = pd.concat([thrice, fours], ignore_index=True)
print(newdf)
```

```python
newdf.to_csv('p_train_upsampled.csv',index=False)
```

```python
three=shuffled.iloc[:half].shape[0]
four=shuffled.iloc[half:].shape[0]
print(three,four)
```

```python
r=range(1,21)
augmentations_3 = [
    [0]+random.sample(r,2) for i in range(three)
]
augmentations_4 = [
    [0]+random.sample(r,3) for i in range(four)
]
augmentations=augmentations_3+augmentations_4
# print(len(np.array(augmentations).flatten()))
```

```python
print(augmentations)
flattened=[]
for i in augmentations:
    for j in i:
        flattened.append(j)
print(3*1200+4*1201)
print(flattened)
print(len(flattened))
```

```python
test=newdf.copy()
test['Augmentation']=flattened
print(test.head())
```

```python
np_train['Augmentation']=[0 for i in range(np_train.shape[0])]
```

```python
np_train.shape[0]
```

```python
final=pd.concat([np_train,test],ignore_index=True).sample(frac=1)
```

```python
print(final)
```

```python
final.to_csv('./train_upsampled.csv',index=False)
```
